CURRENCY CONVERSION CAD TO USD

```python
print(cleaning.dtypes)
```

```python
# convert to float
cleaning['Price Each'] = pd.to_numeric(cleaning['Price Each'], errors='coerce')  # coerce errors to NaN
```

```python
# applying the conversion rate
cad_to_usd = 0.73

# Use dataclean instead of df and ensure 'Price Each' is numeric
cleaning['Price Each'] = cleaning.apply(
    lambda row: float(row['Price Each']) * cad_to_usd if row['Country'] == 'Canada' else row['Price Each'],
    axis=1
)
```

```python
# Convert the 'Order Date' column to datetime, handling errors
cleaning['Order Date'] = pd.to_datetime(cleaning['Order Date'], format='%m/%d/%y %H:%M', errors='coerce')

# Drop rows where conversion failed
cleaning.dropna(subset=['Order Date'])

# Extract year and month
cleaning['Order Year'] = cleaning['Order Date'].dt.year
cleaning['Order Month'] = cleaning['Order Date'].dt.month

# Optional: Extract full month name
cleaning['Order Month Name'] = cleaning['Order Date'].dt.strftime('%B')

# Preview result
cleaning[['Order Date', 'Order Year', 'Order Month', 'Order Month Name']].head()
```

```python
def extract_location(row):
    address = row['Purchase Address']
    country = row['Country']
    try:
        parts = address.split(', ')
        if country == "Canada":
            # Return only the city
            city = parts[1].strip() if len(parts) > 1 else None
            return pd.Series([city])
        elif country == "USA":
            # Return only the city
            city = parts[1].strip() if len(parts) > 1 else None
            return pd.Series([city])
        else:
            return pd.Series([None]) # Return a Series with one None for other countries
    except Exception:
        return pd.Series([None]) # Return a Series with one None on exception

# Drop missing addresses from the *current* dataclean DataFrame
cleaning = cleaning.dropna(subset=['Purchase Address']).copy() # Use .copy() to avoid SettingWithCopyWarning

# Apply the function to the cleaned dataclean DataFrame
cleaning['City'] = cleaning.apply(extract_location, axis=1)
```
